```python
### Extract File From Zip File

import zipfile
zip_ref=zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() # extract file to directry
zip_ref.close() # close file
```

```python
### read data from the file and handle null values

import pandas as pd  
df=pd.read_csv('orders.csv')
df.head(20)

```

```python
df=pd.read_csv('orders.csv' , na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
### rename columns name ... make the lower case and replace space with underscore

##df.rename(columns={'Order Id' : 'order_id' , 'City' :'city'}) --- not a good practice

df.columns
```

```python
df.columns=df.columns.str.lower()

df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')

df.columns
```

```python
### derive new columns discount , sale price and profit

df.head(5)

```

```python
## derive new colums discount , sale price and profit
df['discount']=df['list_price']*df['discount_percent']*.01
df.head()
```

```python
df['sale_price']=df['list_price']-df['discount']
df.head()
```

```python
df['profit']=df['sale_price']-df['cost_price']
df.head()
```

```python
### convert order date from object data type to datetime

df.dtypes
```

```python
### convert order date from object data type to datetime

df['order_date']=pd.to_datetime(df['order_date'] , format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
## drop cost price list price and discount percentent columns

df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
